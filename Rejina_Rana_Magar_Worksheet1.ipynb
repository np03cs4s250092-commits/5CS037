```python
"""Problem - 4: Matrix Operations:
• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:"""
#1. Prove A.A−1 = I.
import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
print(A.dot(np.linalg.inv(A)))
```

```python
#2. Prove AB ̸= BA.
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\nAB:\n", AB)
print("\nBA:\n", BA)
print("\nAre AB and BA equal? ->", np.array_equal(AB, BA))

```

```python
# 3. Prove (AB)^T = B^T * A^T
AB_T = AB.T
BT_AT = np.dot(B.T, A.T)

print("\n(AB)^T:\n", AB_T)
print("\nB^T A^T:\n", BT_AT)
print("\nIs (AB)^T equal to B^T A^T? ->", np.array_equal(AB_T, BT_AT))
```

```python
# 4. Solve the system of equations using inverse method
# System:
# 2x − 3y +  z = −1
#  x −  y + 2z = −3
# 3x +  y −  z =  9

# Matrix form AX = B

A_eq = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3,  1, -1]
])

B_eq = np.array([-1, -3, 9])

# Solve using inverse: X = A^-1 * B
A_eq_inv = np.linalg.inv(A_eq)
X = np.dot(A_eq_inv, B_eq)

print("\nSolution using inverse method (x, y, z):\n", X)
```

```python
# Also solve using numpy built-in solver for verification
X_verify = np.linalg.solve(A_eq, B_eq)

print("\nSolution using np.linalg.solve:\n", X_verify)
```
